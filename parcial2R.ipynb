```python
# Inicializa una lista vacía llamada 'carrito'
carrito = []

# Bucle infinito hasta que el usuario elija salir
while True:
    # Imprime el menú de opciones
    print("\n1. Agregar")
    print("2. Eliminar")
    print("3. Ver")
    print("4. Salir")

    print("---------------------")

    # Solicita al usuario una opción del menú
    opcion = input("Opción: ")

    # Opción 1: Agregar producto al carrito
    if opcion == "1":
        producto = input("Producto: ")  # Pide el nombre del producto
        carrito.append(producto)        # Agrega el producto a la lista
        print("Agregado")               # Confirma al usuario

    # Opción 2: Eliminar producto del carrito
    elif opcion == "2":
        # Si el carrito no está vacío
        if len(carrito) > 0:
            # Muestra los productos enumerados
            for i, p in enumerate(carrito, 1):
                print(f"{i}. {p}")
            # Pide al usuario el número del producto a eliminar
            num = int(input("Número: "))
            # Elimina el producto correspondiente (ajusta por índice base 0)
            carrito.remove(carrito[num - 1])
            print("Eliminado")          # Confirma eliminación
        else:
            print("Vacío")             # Informa si el carrito está vacío

    # Opción 3: Ver productos en el carrito
    elif opcion == "3":
        # Si el carrito tiene productos
        if len(carrito) > 0:
            # Muestra cada producto con guion
            for p in carrito:
                print("-", p)
        else:
            print("Vacío")             # Informa si el carrito está vacío

    # Opción 4: Salir del programa
    elif opcion == "4":
        break                           # Sale del bucle y finaliza

    # Cualquier otra opción no válida
    else:
        print("Opcion NO valida")      # Informa de opción incorrecta
```

```python
# Definimos los roles válidos que se permiten para registrar usuarios
ROLES_VALIDOS = ('admin', 'editor', 'visitante')

# Creamos una lista vacía para almacenar los usuarios registrados
usuarios_registrados = []

# Imprimimos un mensaje inicial de bienvenida
print("Registro de usuarios")

# Mostramos los roles válidos para que el usuario sepa qué opciones puede elegir
print(f"Roles: {ROLES_VALIDOS}\n")

# Iniciamos un bucle infinito para registrar múltiples usuarios
while True:
    # Solicitamos al usuario ingresar un nombre
    # También se puede escribir 'salir' para terminar el registro
    nombre = input("Nombre (o 'salir'): ")

    # Si el usuario escribe 'salir', se rompe el bucle y se termina el registro
    if nombre == 'salir':
        break

    # Solicitamos al usuario que indique el rol
    rol = input("Rol: ")

    # Verificamos si el rol ingresado está dentro de los roles válidos
    if rol in ROLES_VALIDOS:
        # Si es válido, se crea una tupla con el nombre y rol
        usuario = (nombre, rol)
        # Se agrega esa tupla a la lista de usuarios registrados
        usuarios_registrados.append(usuario)
        print("Usuario registrado\n")  # Confirmación al usuario
    else:
        # Si el rol no es válido, se informa al usuario
        print("Rol NO valido\n")

# Al salir del bucle, mostramos la lista de usuarios registrados
print("\nUsuarios registrados:")

# Recorremos la lista e imprimimos cada usuario con su rol
for nombre, rol in usuarios_registrados:
    print(f"- {nombre}: {rol}")
```

```python
# Se crea un diccionario vacío para almacenar los contactos
# Clave = nombre del contacto, Valor = número de teléfono
agenda = {}

# Bucle principal del programa que se ejecuta hasta que el usuario elija salir
while True:
    # Menú principal
    print("\nContactoos")  # (Nota: tiene un error tipográfico, debería ser "Contactos")
    print("1. Agregar contacto")
    print("2. Buscar teléfono")
    print("3. Mostrar contactos")
    print("4. Salir")

    # Se solicita al usuario que elija una opción
    opcion = input("\nOpción: ")

    # Opción 1: Agregar contacto
    if opcion == "1":
        nombre = input("Nombre: ")         # Se pide el nombre del contacto
        telefono = input("Teléfono: ")     # Se pide el número de teléfono
        agenda.update({nombre: telefono})  # Se agrega o actualiza el contacto en el diccionario
        print("Contacto agregado")         # Mensaje de confirmación

    # Opción 2: Buscar un teléfono por nombre
    elif opcion == "2":
        nombre = input("Nombre a buscar: ")  # Se pide el nombre a buscar
        telefono = agenda.get(nombre)        # Se intenta obtener el teléfono del diccionario

        if telefono:
            print(f"Teléfono: {telefono}")   # Si se encuentra, se muestra
        else:
            print("Contacto no encontrado")  # Si no se encuentra, se informa al usuario

    # Opción 3: Mostrar todos los contactos
    elif opcion == "3":
        if agenda:
            print("\nContactos:")                         # Título
            for nombre, telefono in agenda.items():       # Se recorre el diccionario
                print(f"{nombre}: {telefono}")            # Se imprime cada contacto
        else:
            print("Agenda vacía")                         # Si no hay contactos

    # Opción 4: Salir del programa
    elif opcion == "4":
        print("Hasta luego")  # Mensaje de despedida
        break                 # Se rompe el bucle y finaliza el programa

    # Si la opción ingresada no es válida
    else:
        print("Opción NO valida")  # Mensaje de error
```

```python
# 4.

votosCo = {
    "rojo": 10,
    "negro": 21,
    "lila": 12,
    "amarillo": 0
}

voto_usu = input("Vota por un color").lower()

# Verificar si el color está en las opciones
if voto_usu in votosCo:
    # Actualizar el valor (sumar un voto)
    votosCo[voto_usu] += 1
    print(f"\n¡Voto registrado! votaste por {voto_usu}")
else:
    print(f"\nOpción NO valida. '{voto_usu}' no está en la lista de colores disponibles.")

# Mostrar resultados actuales
print("\n--- Resultados actuales ---")
for color, votos in votosCo.items():
    print(f"{color.capitalize()}: {votos} voto(s)")
```
